# Selektion
Kontrollstrukturen lassen uns entscheiden ob man einen oder mehrere Befehle ausführen soll oder nicht und ob man einen Codeblock wiederholen muss oder nicht. Wir schauen uns diese Kontrollstrukturen an und schreiben zur Übung kleine Programme dazu.
## Bedingungen
Mit Bedingungen steuern wir, ob ein Befehlsblock ausgeführt wird oder nicht. Eine Bedingung ist dabei ein Ausdruck, den der Computer entscheiden kann ob er wahr (True) oder falsch (False) ist.
~~~ python
a == 3        # zwei Gleichheitszeichen damit es keine Zuordnung wird!
3 <= 4
a > 5
a != 'ja'
'e' in a
~~~
## If-Befehl
Mit dem Schlüsselwort if wird dieser Befehl begonnen. Im Anschluss muss eine Bedingung fol-gen. Mit dem Doppelpunkt wird der Codeblock begonnen, der ausgeführt wird falls die Bedin-gung erfüllt ist. Dieser Codeblock muss eingerückt sein gegenüber dem Schlüsselwort if.
Ist der Codeblock beendet schreibt man wieder auf der gleichen Höhe wie if. Man kann einen weiteren else: Befehl angeben. Dieser Codeblock wird ausgeführt falls die erste Bedingung nicht erfüllt ist.
~~~ python
if a ==3 :
    # Codeblock falls die Bedingung erfüllt ist.
    print(' a ist 3')
else:
    # Codeblock falls die Bedingung nicht erfüllt ist.
    print('a ist nicht 3')
print('dieser Befehl wird sicher ausgeführt')
~~~
Bitte beachten Sie, dass jeder Codeblock etwas enthalten muss. Den else-Teil des Befehles kann man ganz weglassen. Sollte man das else aber schreiben, muss mindestens ein Platzhalter vor-handen sein.
~~~ python
if a ==3 :
    # Codeblock falls die Bedingung erfüllt ist.
    print(' a ist 3')
else:
    pass # Codeblock ist leer.
print('dieser Befehl wird sicher ausgeführt')
~~~
oder 
~~~ python
if a ==3 :
    # Codeblock falls die Bedingung erfüllt ist.
    print(' a ist 3')
print('dieser Befehl wird sicher ausgeführt')
~~~
### Beispiel
Schreiben Sie ein Programm in welchem man eine zufällige Zahl herausfinden muss. Der Computer erzeugt eine geheime Zahl zwischen 1 und 100.
~~~ python
import random

geheimzahl = random.randrange(0, 100)
~~~
Der erste Befehl lädt die Befehlsbibliothek random. Der Importbefehl sollte ganz am Anfang eines Programmes stehen bevor irgendein Befehl sonst geschrieben wird. In dieser Bibliothek sind alle Befehle programmiert, die man für die Erzeugung einer Zufallszahl braucht.
Der Benutzer soll nun raten bis er die richtige Zahl gefunden hat. Das Programm zählt mit, wie viele Versuche der Benutzer braucht und gibt bei jedem Versuch ein Feedback «zu hoch gera-ten», «zu tief geraten» aus.
### Ablauf des Programmes
1. Laden der Bibliothek
2. Erzeugen der Geheimzahl, Zähler auf null setzen.
3. Benutzer nach einer Zahl fragen
4. Vergleichen und Feedback «zu gross», «zu tief» geben
5. Zähler wird um eins erhöht.
6. Richtigen Zahl geraten? Ja, weiter mit Schritt 7 ansonsten zurück zu Nummer 3
7. «Sie haben es geschafft» ausgeben.


```python
import random

geheimzahl = random.randrange(0, 100)
guess = -1  #startwert der sicherlich falsch ist.
zaehler =0
print("Sie müssen eine Zahl zwischen 0 und 100 raten.")
while guess != geheimzahl:
    guess = int(input("Geben Sie ihren Tipp ab: "))
    if guess > geheimzahl:
        print("Sie haben zu hoch geraten")
    elif guess < geheimzahl:
        print("Sie haben zu tief geraten")
    else:
        print("Sie haben die Zahl gefunden!!")
    zaehler += 1

print("Sie mussten {} Mal raten!".format(zaehler))
```

## Testen von Primzahlen
Wie kann man testen ob einen gegebene natürliche Zahl einee Primzahl ist oder nicht?

```python
# Testen von Primzahlen
def isItPrime(n=3):
    if n<2:
        return False
    else:
        i=2
        while i**2<=n:
            if n%i==0:
                return False
            i=i+1
    return True


n = int(input("Geben Sie eine ganze Zahl ein: "))
if isItPrime(n):
    erg = "Ja"
else:
    erg = "Nein"
print("is it prime: {}".format(erg))
```

# Collatz-Conjecture

Das Programm soll weiterarbeiten falls n>1 ist. Ist die Zahl eine ganze Zahl (`n%2==0`) so soll sie durch zwei dividiert werden (`n//2`). Ist die Zahl aber ungerade, so soll die Zahl mit drei multipliziert werden und eins addiert. (`n = 3*n+1`)

Schreiben sie ein Programm dass alle Zahlen ausgibt und bei eins aufhört.

```python
def collatz(n=3):
    while n>1:
        if n%2==0:
            n=n//2
        else:
            n= 3*n+1
        print(n)
    return print("Habe 1 erreicht!")

def collatzNum(n=3):
    i=0
    while n>1:
        if n%2==0:
            n=n//2
        else:
            n= 3*n+1
        i=i+1
    return print("Anzahl durchläufe: {}".format(i))

```
